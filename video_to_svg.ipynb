Create PNGs

```python
import cv2

vid = cv2.VideoCapture("bad-apple.mp4")
ret, image = vid.read()

frame_num = 0
while ret:
    # write as pngs to frames directory
    if frame_num % 10 == 0 or frame_num == 0 or frame_num == (vid.get(cv2.CAP_PROP_FRAME_COUNT) - 1):
        print(f"frame_num = {frame_num}")
    # add a 1 pixel black border to the outside
    image = cv2.copyMakeBorder(image, 1, 1, 1, 1, cv2.BORDER_CONSTANT, value=[0, 0, 0])
    cv2.imwrite(f"frames_border/{frame_num}.png", image)
    ret, image = vid.read()
    frame_num += 1
```

Create Bitmaps

```python
from PIL import Image
from pathlib import Path

Path("bitmaps_border").mkdir(exist_ok=True)

for i in range(6572):
    img = Image.open(f"frames_border/{i}.png")
    img.save(f"bitmaps_border/{i}.bmp")
```

```python
import os
svg_path = Path("svgs_border")
if not svg_path.exists():
    svg_path.mkdir()
    
for i in range(6572):
    os.system(f"potrace -o svgs_border/{i}.svg -b svg bitmaps_border/{i}.bmp")
    print(i)
```

```python


for i in range(6572):
    svg_file = f"svgs/{i}.svg"
    with open(svg_file, "r") as f:
        svg_content = f.read()

# Add transparent rectangle to SVG content
    svg_content = svg_content.replace(
        '</svg>',
        '<rect x="0" y="0" width="960" height="720" fill="none" stroke="none" />\n</svg>'
    )
    modified_path = f"svgs_border_2/{i}.svg"
# Write modified SVG content to a new file
    with open(modified_path, "w") as f:
        f.write(svg_content)
```

```python
# calculate the center of mass of every frame

import json
coms = {}
for i in range(6572):
    img = cv2.imread(f"frames_border/{i}.png")
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    # find com by taking average of black pixels
    x = 0
    y = 0
    count = 0
    for _i in range(thresh.shape[0]):
        for j in range(thresh.shape[1]):
            if thresh[_i, j] == 0:
                x += j
                y += _i
                count += 1
    x /= count
    y /= count
    # i dunno why its off by 0.5
    coms[i] = (x - 0.5, y - 0.5)
    print((x - 0.5, y - 0.5))

with open("centerofmasses.json", "w") as f:
    json.dump(coms, f)
    
    
```
